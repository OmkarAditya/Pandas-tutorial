## Manipulating Data


```python
#all lower case (mark actual column values not changed..)

car_sales["Make"].str.lower()
```

```python
#uppercase  (to change actual column values inside car_sales..need to do dis)..
#reassignment

car_sales["Make"]=car_sales["Make"].str.lower()
car_sales
```

```python
#data not complete
car_sales_missing=pd.read_csv("car-sales-missing-data.csv")
car_sales_missing  #mark the NaN columns
```

```python
#fillna funct to fill NaN values

car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean())
```

```python
# inplace 

# we can do reassigning by a[x]= a[x].fillna().... OR  use inplace
# inplace = true means it updates in the actual dataframe

car_sales_missing["Odometer"].fillna(car_sales_missing["Odometer"].mean(),inplace=True)
car_sales_missing
```

```python
#dropna function removes all columns with NaN values

car_sales_missing.dropna(inplace=True)
car_sales_missing
```

```python
# lambda function

# importing pandas library
import pandas as pd
  
# creating and initializing a list
values= [['Rohan',455],['Elvish',250],['Deepak',495],
         ['Soni',400],['Radhika',350],['Vansh',450]]
 
# creating a pandas dataframe
df = pd.DataFrame(values,columns=['Name','Total_Marks'])
 
# Applying lambda function to find
# percentage of 'Total_Marks' column
# using df.assign()
df = df.assign(Percentage = lambda x: (x['Total_Marks'] /500 * 100))
 
# displaying the data frame
df
```

```python
# deleting column using del
del df['Percentage']
df
```

```python
df_new = df.reindex([1,2,4,5,3])
df_new
```
